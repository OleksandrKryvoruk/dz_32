**Розв'язок:**

```python
import pandas as pd
import numpy as np

students_list = ["Leonardo DiCaprio",
                 "Matt Damon",
                 "Jack Nicholson",
                 "Mark Wahingerg",
                 "Martin Sheen",
                 "Ray Winstone",
                 "Vera Farmiga",
                 "Alec Baldwin",
                 "Anthony Anderson",
                 "Kevin Corrigan",
                 "James Badge Dale",
                 "David O'Hara",
                 "Mark Rolston",
                 "Robert Wahlberg",
                 "Kristen Dalton"]

df = pd.DataFrame(students_list, columns=['name'])                           #1_а). Iм'я та прізвище студента
df['aver_score'] = np.random.normal(8.0, 1.5, 15)                            #1_б). Середній бал
df['ex_score'] = [np.random.randint(1, 12) for _ in df['name']]              #1_в). Оцінка за поточний екзамен
df['count_tries'] = [np.random.randint(1, 3) for _ in df['name']]            #1_г). Кількість спроб по поточному екзамену
df['trans_to_kourse'] = [True if i > 6 else False for i in df['aver_score']] #2. Допуск студента на наступний курс
df
```

```python
#4. Кількість студентів, що здали екзамен з першого разу
print(f" Кількість студентів, що здали поточний екзамен з 1 разу : {len(df.loc[df['count_tries'] == 1])}") 
```

```python
#5. Збережіть будь які дані у вигляді серії
series = pd.Series(df['aver_score'])                                         
series
```

```python
#3. Виведіть max, min, avg, median значення середнього балу студентів
df_data = series.agg(['min', 'max', 'mean', 'median'])
df_data
```
